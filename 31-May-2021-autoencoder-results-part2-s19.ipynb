## This page shows the test results of an autoencoder where the testing waveform is from Radice, et al, 2019

```python
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
```

## The following are the results for 3000 test samples where the time shift, phase shift, amplitude scale factors and the waveform have not been used to train the autoencoder.

### To quantify how good the reconstruction of a waveform is, a quantity, referred to as the correlation factor (CF), defined below, is used,

### $C = \frac{(\boldsymbol{h}^{r}, \boldsymbol{h}^{t})}{\sqrt{(\boldsymbol{h}^{r} , \boldsymbol{h}^{r}) (\boldsymbol{h}^{t}, \boldsymbol{h}^{t})}}$,
### where $C$ is the CF, $h^{r}$ is the reconstruted waveform from the trained autoencoder in vector form, and $h^{t}$ the target waveform in vector form.
### Both $h^{r}$ and $h^{t}$ can either be in the time (T) domain or the time-frequency (TF) domain,
### and $(,)$ is the scalar product between two vectors, given by
### $(\boldsymbol{a}, \boldsymbol{b}) = \sum_n a_n b_n$.
### The CF is used both in the time domain and the time-frequency domain.
### Larger value of CF means closer matching between $h^{r}$ and $h^{t}$.


## Figure 1.  The distribution of CF in the TF domain for the 3000 test samples.
## Left panel : the distribution of CF between $h^{r}$ and $h^{t}$  in the TF domain.
## Right panel : the distribution of CF between the  $h^{r}$ and random values following Gaussian distribution. 


```python
fig = plt.figure(figsize=(43.5,20.25), dpi= 100, facecolor='w', edgecolor='k')
a=fig.add_subplot(1, 2, 1)
x = Image.open('s19/cf_dist_in_tf_and_random.png') 
plt.imshow(x)
plt.axis('off')

plt.show()
```
