## This page shows the test results of an autoencoder where the test samples have not been used to train the autoencoder

## The waveform is from Radice, et al, 2019. 
## The progenitor is of 12 M$_{\odot}$, no rotation (s12 from Sukhbold, et al, 2016), equatorial observer.
## Equation of state : SFHo (Steiner, et al, 2013)

```python
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
```

## The following are the results for 2800 testing samples where the testing samples (inputs and target outputs) have not been used during training

## To quantify how good the reconstruction of a waveform is, a quantity, referred to as the correlation factor, defined below, was used,

## $C = \frac{(h^{r}, h^{t})}{\sqrt{(h^{r} , h^{r}) (h^{t}, h^{t})}}$
## where C is the correlation factor, $h^{r}$ is the reconstruted waveform from the trained autoencoder, and $h^{t}$ the target waveform 
## and $(,)$ is the scalar product between two waveforms, given by
## $(a, b) = \int a(t)b(t)dt$

## The range for this value is [-1, 1]. 
## a correlation factor of 1 means a perfect matching, -1 opposite matching


## Figure 1.  The distribution of the correlation factors for the 2800 unseen test samples


```python
fig = plt.figure(figsize=(25.5,12.25), dpi= 100, facecolor='w', edgecolor='k')
a=fig.add_subplot(1, 2, 1)
x = Image.open('03-Mar-2021/cf_in_td_not_seen.png') 
plt.imshow(x)
plt.axis('off')

plt.show()
```
