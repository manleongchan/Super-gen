## This page aims to show the problem of reconstructing a time shifted waveform using an autoencoder where the input is the corresponding value of time shift.

```python
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
```

# The following shows the results from an autoencoder where the input is either a time shifted waveform or just a single time shift value

## Figure 1. Training loss and validation loss as a function of epochs. 
##          The input to the autoencoder is a time shifted sine-Gaussian waveform.
##          The output is supposed to be the same waveform.
##          The encoder is a convolutional neural network.

```python
fig = plt.figure(figsize=(43.5,20.25), dpi= 100, facecolor='w', edgecolor='k')
a=fig.add_subplot(1, 2, 1)
x = Image.open('../t4/seed_2000_300epochs_network_loss.png') 
plt.imshow(x)
plt.axis('off')

plt.show()
```

## Figure 2. an exmple output from the trained autoencoder and the input

```python
fig = plt.figure(figsize=(43.5,20.25), dpi= 100, facecolor='w', edgecolor='k')
a=fig.add_subplot(1, 2, 1)
x = Image.open('../t4/waveform_as_input.png') 
plt.imshow(x)
plt.axis('off')

plt.show()
```

## Figure 3. Training loss and validation loss as a function of epochs. 
##          The input to the autoencoder is a single value representing a time shift (the range for time shift is randomly drawn from (-0.9, 0.9), the values are normalised before being input to the autoencoder).
##          The output is supposed to be a sine-Gaussian waveform shifted in time by the input.
##          The encoder could be neural network of only dense layers, convolutional neural network, or neural network of only dense layers followed by a convolutional neural network. The results are similar.
##          Every thing else is the same as the autoencoder above

```python
fig = plt.figure(figsize=(43.5,20.25), dpi= 100, facecolor='w', edgecolor='k')
a=fig.add_subplot(1, 2, 1)
x = Image.open('../t4/seed_10000_300epochs_nn_network_loss.png') 
plt.imshow(x)
plt.axis('off')

plt.show()
```
